```python
#JAWABAN SOAL NO.1

def f(x):
  return x**2 - x - 2
```

```python
#(1)BISECTION
def bisection(f, a, b, tol=1e-6, max_iter=100):

  for i in range(max_iter):
    c = (a + b) / 2

    if abs(f(c)) < tol:
      return c

    if f(a) * f(c) < 0:
      b = c
    else:
      a = c

    return c
```

```python
#(2)REGULAFALSI
def regulafalsi(f, a, b, tol=1e-6, max_iter=100):

  for i in range(max_iter):

    c = b - (f(b)*(b-a))/(f(b)-f(a))

    if abs(f(c)) < tol:
      return c

    if f(a) * f(c) < 0:
      b = c
    else:
      a = c

    return c
```

```python
#(3)SECANT
def secant(f, x0, x1, tol=1e-6, max_iter=100):

  for i in range(max_iter):
    x2 = x1 - f(x1)*(x1-x0)/(f(x1)-f(x0))

    if abs(f(x2)) < tol:
      return x2

    x0 = x1
    x1 = x2

    return x2
```

```python
def df(x):
  return 2*x - 1
```

```python
#(4)NEWTONRAP
def newtonrap(f, df, x0, tol=1e-6, max_iter=100):
  for i in range(max_iter):
    x1 = x0 - f(x0)/df(x0)

    if abs(f(x1)) < tol:
      return x1

      x0 = x1

    return x1
```

```python
#JAWABAN SOAL NO.2

# Bisection
root_bisection = bisection(f, 1, 3)
print("Akar dengan metode Bisection:", root_bisection)

# Regula Falsi
root_regulafalsi = regulafalsi(f, 1, 2)
print("Akar dengan metode Regula Falsi:", root_regulafalsi)

# Secant
root_secant = secant(f, 1, 2)
print("Akar dengan metode Secant:", root_secant)

# Newton Raphson
root_newton = newtonrap(f, df, 2)
print("Akar dengan metode Newton Raphson:", root_newton)
```

```python
#JAWABAN SOAL NO.3

def regulafalsi_iterasi(f, a, b, tol=1e-6, max_iter=20):

    print("Iterasi    a           b           c           f(c)")
    print("-------------------------------------------------------")

    for i in range(1, max_iter+1):

        c = b - (f(b)*(b-a))/(f(b)-f(a))

        print(f"{i:<10}{a:<12.6f}{b:<12.6f}{c:<12.6f}{f(c):<12.6f}")

        if abs(f(c)) < tol:
            break

        if f(a)*f(c) < 0:
            b = c
        else:
            a = c

    print("\nSolusi akar pendekatan:", c)
```

```python
regulafalsi_iterasi(f, 1, 3)
```

```python
def secant_iterasi(f, x0, x1, tol=1e-6, max_iter=20):

    print("Iterasi    x_n         f(x_n)        Error")
    print("------------------------------------------------")

    for i in range(1, max_iter+1):

        x2 = x1 - f(x1)*(x1-x0)/(f(x1)-f(x0))
        error = abs(x2-x1)

        print(f"{i:<10}{x1:<12.6f}{f(x1):<12.6f}{error:<12.6f}")

        if error < tol:
            break

        x0 = x1
        x1 = x2

    print("\nAkar aproksimasi:", x2)
```

```python
# JAWABAN Soal No.4

# Masalah fisika: Sebuah mobil bergerak dengan percepatan berubah sehingga posisinya mengikuti persamaan di bawah. Tentukan waktu saat posisi benda = 0 (berada tepat di titik acuan)
# Persamaan posisi: s(t) = t^2 - 4t + 3
# Dicari akar dari s(t) = 0
# Kasus ini dipecahkan dengan menggunakan metode Secant
```

```python
def f(t):
  return t**2 - 4*t +3
```

```python
def secant(f, x0, x1, tol=1e-6, max_iter=100):

    for i in range(max_iter):

        x2 = x1 - f(x1)*(x1-x0)/(f(x1)-f(x0))

        if abs(x2 - x1) < tol:
            return x2

        x0 = x1
        x1 = x2

    return x2
```

```python
root = secant(f, 1, 4)
print("Waktu saat posisi = 0:", root)
```

```python
# Penjelasan:
# Persamaan memiliki dua akar yaitu t = 1 dan t = 3
# Metode Secant menghasilkan t = 1.0
# Nilai tersebut merupakan salah satu akar dari persamaan tersebut
```

```python
#JAWABAN SOAL NO.5

import math

def g(x):
  return math.sqrt(x + 2)
```

```python
def fixed_point(g, x0, tol=1e-6, max_iter=20):

    print("Iterasi   x_n         Error")
    print("--------------------------------")

    for i in range(1, max_iter + 1):

        x1 = g(x0)
        error = abs(x1 - x0)

        print(f"{i:<10}{x1:<12.6f}{error:<12.6f}")

        if error < tol:
            break

        x0 = x1

    print("\nAkar aproksimasi:", x1)
```

```python
fixed_point(g, 1)
```

```python
# Metode Fixed Point Iteration digunakan untuk mencari akar persamaan non-linear
# Persamaan f(x) = x^2 - x - 2 diubah menjadi x = sqrt(x + 2)
# Iterasi dilakukan hingga nilai konvergen
# Hasil menunjukkan akar mendekati x = 2
```
