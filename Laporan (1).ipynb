# 1. Pendahuluan

Gerak jatuh bebas merupakan salah satu fenomena dasar dalam fisika yang dipengaruhi oleh gaya gravitasi bumi. Pada kondisi ideal tanpa hambatan udara, benda akan mengalami percepatan konstan akibat gravitasi. Namun, pada kondisi nyata, gerak benda juga dipengaruhi oleh gaya hambat udara yang bekerja berlawanan arah dengan gerak benda. Besarnya gaya hambat udara umumnya bergantung pada kecepatan benda sehingga menyebabkan gerak benda menjadi lebih kompleks dibandingkan gerak jatuh bebas ideal.

Ketika benda terus bergerak turun, gaya hambat udara akan semakin besar hingga mencapai kondisi setimbang dengan gaya gravitasi. Pada kondisi tersebut, resultan gaya bernilai nol sehingga benda bergerak dengan kecepatan konstan yang disebut sebagai kecepatan terminal. Persamaan yang diperoleh pada kasus ini berbentuk persamaan nonlinear sehingga penyelesaiannya memerlukan metode numerik. Oleh karena itu, pada studi kasus ini digunakan metode Newton-Raphson untuk menentukan nilai kecepatan terminal secara iteratif.

# 2. Metodologi

Pada studi kasus ini dianalisis sebuah benda bermassa 5 kg yang jatuh di udara dengan pengaruh gaya hambat udara. Gaya-gaya yang bekerja pada benda terdiri atas gaya gravitasi dan gaya hambat udara.

Gaya gravitasi:

$$
F_g = mg
$$

Gaya hambat udara:

$$
F_d = kv^2
$$

Saat kecepatan terminal tercapai:

$$
mg-kv^2=0
$$

Dengan:
- m = massa benda (kg)
- g = percepatan gravitasi (m/s²)
- k = konstanta hambatan udara
- v = kecepatan benda (m/s)

Parameter yang digunakan:
- massa benda = 5 kg
- gravitasi = 9.8 m/s²
- konstanta hambatan = 0.25

Sehingga diperoleh persamaan nonlinear:

$$
f(v)=49-0.25v^2
$$

Untuk menyelesaikan persamaan nonlinear tersebut digunakan metode Newton-Raphson. Persamaan umum metode Newton-Raphson adalah:

$$
x_{n+1}=x_n-\frac{f(x_n)}{f'(x_n)}
$$

Turunan fungsi:

$$
f'(v)=-0.5v
$$

# Perhitungan Iterasi Metode Newton-Raphson

Persamaan nonlinear yang digunakan:

$$
f(v)=49-0.25v^2
$$

Turunan fungsi:

$$
f'(v)=-0.5v
$$

Rumus Newton-Raphson:

$$
v_{n+1}=v_n-\frac{f(v_n)}{f'(v_n)}
$$

Tebakan awal:

$$
v_0=10
$$

---

## Iterasi 1

Hitung nilai fungsi:

$$
f(10)=49-0.25(10)^2
$$

$$
f(10)=49-25
$$

$$
f(10)=24
$$

Hitung turunan fungsi:

$$
f'(10)=-0.5(10)
$$

$$
f'(10)=-5
$$

Substitusi ke rumus Newton-Raphson:

$$
v_1=10-\frac{24}{-5}
$$

$$
v_1=10+4.8
$$

$$
v_1=14.8
$$

Error:

$$
|14.8-10|=4.8
$$

---

## Iterasi 2

Hitung nilai fungsi:

$$
f(14.8)=49-0.25(14.8)^2
$$

$$
f(14.8)=49-54.76
$$

$$
f(14.8)=-5.76
$$

Hitung turunan fungsi:

$$
f'(14.8)=-0.5(14.8)
$$

$$
f'(14.8)=-7.4
$$

Substitusi ke rumus Newton-Raphson:

$$
v_2=14.8-\frac{-5.76}{-7.4}
$$

$$
v_2=14.8-0.778378
$$

$$
v_2=14.021622
$$

Error:

$$
|14.021622-14.8|=0.778378
$$

---

## Iterasi 3

Hitung nilai fungsi:

$$
f(14.021622)=49-0.25(14.021622)^2
$$

$$
f(14.021622)\approx -0.151
$$

Hitung turunan fungsi:

$$
f'(14.021622)=-0.5(14.021622)
$$

$$
f'(14.021622)\approx -7.010811
$$

Substitusi ke rumus Newton-Raphson:

$$
v_3=14.021622-\frac{-0.151}{-7.010811}
$$

$$
v_3\approx14.000017
$$

Error:

$$
|14.000017-14.021622|=0.021605
$$

---

## Iterasi 4

Hitung nilai fungsi:

$$
f(14.000017)\approx0
$$

Substitusi ke rumus Newton-Raphson:

$$
v_4\approx14.000000
$$

Error:

$$
|14.000000-14.000017|=0.000017
$$

---

## Hasil Akhir

Diperoleh kecepatan terminal:

$$
v \approx 14 \, m/s
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# PARAMETER FISIKA
m = 5
g = 9.8
k = 0.25

# FUNGSI NONLINEAR
def f(v):
    return m*g - k*v**2

# TURUNAN FUNGSI
def df(v):
    return -2*k*v

# METODE NEWTON-RAPHSON
v = 10
toleransi = 0.0001
maks_iterasi = 20

iterasi_list = []
kecepatan_list = []
error_list = []

print("METODE NEWTON-RAPHSON")
print("-" * 40)

for i in range(maks_iterasi):

    v_baru = v - f(v)/df(v)

    error = abs(v_baru - v)

    iterasi_list.append(i+1)
    kecepatan_list.append(v_baru)
    error_list.append(error)

    print(f"Iterasi ke-{i+1}")
    print(f"v = {v_baru:.6f} m/s")
    print(f"Error = {error:.6f}")
    print("-" * 40)

    if error < toleransi:
        break

    v = v_baru

print("\nHASIL AKHIR")
print(f"Kecepatan terminal = {v_baru:.6f} m/s")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.plot(iterasi_list, kecepatan_list, marker='o')

plt.title("Grafik Konvergensi Newton-Raphson")
plt.xlabel("Iterasi")
plt.ylabel("Kecepatan (m/s)")
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.plot(iterasi_list, error_list, marker='o')

plt.title("Grafik Error Iterasi")
plt.xlabel("Iterasi")
plt.ylabel("Error")
plt.grid(True)

plt.show()
```

# 3. Hasil dan Pembahasan

Berdasarkan hasil iterasi metode Newton-Raphson diperoleh nilai kecepatan terminal sebesar 14 m/s. Nilai tersebut diperoleh setelah empat iterasi dengan error yang semakin kecil pada setiap iterasi.

Metode Newton-Raphson menunjukkan proses konvergensi yang cepat. Hal ini terlihat dari perubahan nilai error yang menurun secara signifikan dari iterasi pertama hingga iterasi terakhir.

Secara fisika, kecepatan terminal terjadi ketika gaya gravitasi sama dengan gaya hambat udara sehingga resultan gaya menjadi nol. Pada kondisi tersebut benda tidak lagi mengalami percepatan dan bergerak dengan kecepatan konstan.

Grafik konvergensi menunjukkan bahwa nilai solusi semakin stabil menuju 14 m/s. Selain itu, grafik error memperlihatkan bahwa selisih antar iterasi semakin mendekati nol sehingga solusi dapat dikatakan konvergen.
