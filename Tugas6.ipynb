```python
#Soal 1

import numpy as np

# Fungsi gaya
def F(x):
    return 5 + 2*x**2

# Metode Midpoint
def midpoint(f, a, b, n):
    dx = (b - a)/n
    titik_tengah = a + (np.arange(n) + 0.5)*dx
    return dx * np.sum(f(titik_tengah))

# Perhitungan n = 4
W_n4 = midpoint(F, 0, 4, 4)
print("Hasil Midpoint (n=4) =", W_n4)

# Konvergensi
tol = 1e-14
n = 4
I_lama = midpoint(F, 0, 4, n)

while True:
    n *= 2
    I_baru = midpoint(F, 0, 4, n)

    if abs(I_baru - I_lama) < tol:
        break

    I_lama = I_baru

print("n optimal =", n)
print("Hasil konvergen =", I_baru)
```

```python
#Soal 2

import numpy as np

# Fungsi kapasitas kalor
def C(T):
    return 20 + 0.05*T**2

# Metode Trapesium
def trapezium(f, a, b, n):
    h = (b-a)/n
    x = np.linspace(a, b, n+1)

    return h/2 * (
        f(x[0])
        + 2*np.sum(f(x[1:n]))
        + f(x[n])
    )

# Perhitungan n = 4
Q_n4 = trapezium(C, 0, 100, 4)
print("Hasil Trapesium (n=4) =", Q_n4)

# Konvergensi
tol = 1e-14
n = 4
I_lama = trapezium(C, 0, 100, n)

while True:
    n *= 2
    I_baru = trapezium(C, 0, 100, n)

    if abs(I_baru - I_lama) < tol:
        break

    I_lama = I_baru

print("n optimal =", n)
print("Hasil konvergen =", I_baru)
```

```python
#Soal 3

import numpy as np

# Fungsi arus
def I(t):
    return 2 + 3*t + t**2

# Metode Simpson 1/3
def simpson13(f, a, b, n):

    if n % 2 != 0:
        raise ValueError("n harus genap")

    h = (b-a)/n
    x = np.linspace(a, b, n+1)

    hasil = (
        f(x[0])
        + f(x[-1])
        + 4*np.sum(f(x[1:-1:2]))
        + 2*np.sum(f(x[2:-1:2]))
    )

    return h/3 * hasil

# Perhitungan n = 6
q_n6 = simpson13(I, 0, 6, 6)
print("Hasil Simpson 1/3 (n=6) =", q_n6)

# Konvergensi
tol = 1e-14
n = 6
I_lama = simpson13(I, 0, 6, n)

while True:
    n *= 2

    if n % 2 != 0:
        n += 1

    I_baru = simpson13(I, 0, 6, n)

    if abs(I_baru - I_lama) < tol:
        break

    I_lama = I_baru

print("n optimal =", n)
print("Hasil konvergen =", I_baru)
```
