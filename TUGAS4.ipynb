Suatu rangkaian listrik terdiri dari empat loop dengan arus masing-masing:

I1, I2, I3, I4

Berdasarkan Hukum Kirchhoff II (KVL), diperoleh sistem persamaan berikut:

10I1 - 2I2 - I3 - I4 = 3

-2I1 + 10I2 - I3 - I4 = 15

-I1 - I2 + 10I3 - 2I4 = 27

-I1 - I2 - 2I3 + 10I4 = -9

1. Nyatakan sistem persamaan di atas dalam bentuk matriks Ax = b

2. Tentukan nilai arus:
I1, I2, I3, I4

Dengan metode:

metode dekomposisi LU

metode Jacobi

metode Gauss-Seidel

```python
# metode dekomposisi LU

import numpy as np
from scipy.linalg import lu

A = np.array([
    [10, -2, -1, -1],
    [-2, 10, -1, -1],
    [-1, -1, 10, -2],
    [-1, -1, -2, 10]
], dtype=float)

b = np.array([3, 15, 27, -9], dtype=float)

P, L, U = lu(A)

# Solve Ly = Pb
Pb = np.dot(P, b)
y = np.linalg.solve(L, Pb)

# Solve Ux = y
x = np.linalg.solve(U, y)

print("Solusi LU:", x)
```

```python
#Codingan Jacobi

def jacobi(A, b, x0, tol=1e-6, max_iter=100):
    n = len(b)
    x = x0.copy()

    for _ in range(max_iter):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j]*x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s)/A[i][i]

        if np.linalg.norm(x_new - x) < tol:
            return x_new
        x = x_new

    return x

x0 = np.zeros(4)
print("Jacobi:", jacobi(A, b, x0))
```

```python
#Metode gauss seidel

def gauss_seidel(A, b, x0, tol=1e-6, max_iter=100):
    n = len(b)
    x = x0.copy()

    for _ in range(max_iter):
        x_old = x.copy()

        for i in range(n):
            s1 = sum(A[i][j]*x[j] for j in range(i))
            s2 = sum(A[i][j]*x_old[j] for j in range(i+1, n))
            x[i] = (b[i] - s1 - s2)/A[i][i]

        if np.linalg.norm(x - x_old) < tol:
            return x

    return x

print("Gauss-Seidel:", gauss_seidel(A, b, x0))
```
